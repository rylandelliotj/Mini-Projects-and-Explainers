```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
useful_life = 100 # How many years you can use the object
capital_cost = 10000 # Initial Cost
years = np.arange(useful_life + 1, step=round(useful_life / 10))
k = 0.01 # k value, or depreciation rate. 
'''
Close together when k = high and useful_life =
Further apart when k = high and useful_life =
'''
```

```python
# Linear Depreciation
'''
Used when the value of an asset is consistent over time.
Example: Well built furniture
'''

def linear(cost, salvage, useful_life):
    return (cost - salvage) / useful_life

salvage = 1000 # How much you can recover at end-of-life
depreciation = linear(capital_cost, salvage, useful_life) # Decrease in value per year

lin_values = capital_cost - depreciation * years
plt.plot(years, lin_values, marker='o')
plt.grid(True)
plt.show()
```

```python
# Geometric Depreciation
'''

'''

def geometric(initial_value, depreciation_rate, time):
    return initial_value * (1 - depreciation_rate) ** time

geo_values = geometric(capital_cost, k, years)

plt.plot(years, geo_values, marker='o')
plt.grid(True)
plt.show()
```

```python
# Hyperbolic Depreciation
'''

'''

def hyperbolic(capital_cost, k, time):
    return capital_cost / (1 + k*time)

hyper_values = hyperbolic(capital_cost, k, years)

plt.plot(years, hyper_values, marker='o')
plt.grid(True)
plt.show()
```

```python
# Plotting both
plt.figure(figsize=(10, 6))
plt.plot(years, geo_values, marker='o', linestyle='-', color='b', label='Geometric Depreciation')
plt.plot(years, hyper_values, marker='o', linestyle='-', color='r', label='Hyperbolic Depreciation')
plt.xlabel('Years')
plt.ylabel('Asset Value')
plt.title('Geometric vs Hyperbolic Depreciation (Same Rate)')
plt.legend()
plt.grid(True)
plt.ylim(0, capital_cost *1.1)
plt.show()
```
